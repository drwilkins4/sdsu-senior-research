Newtonian and the Relativistic calculations
- The MvsR files represent mass as a function of stellar radius.
- The zvsM files represent gravitational redshift as a function of stellar radius.


```python
# Importing numpy and matplotlib
import numpy as np 
import matplotlib.pyplot as plt
```

### Relativistic Calculations both, $M$ vs. $R$

```python
plt.style.use('fivethirtyeight')

## GR Calculations, M vs. R


# linear interpolation used
file_name_2 = 'MvsR_phase1.dat'
all_data_2 = np.loadtxt(file_name_2,skiprows=2)
all_data_2 = all_data_2.transpose()
radius_2 = all_data_2[0]
mass_2 = all_data_2[1]

# linear interpolation used
file_name_3 = 'MvsR_new.dat'
all_data_3 = np.loadtxt(file_name_3,skiprows=2)
all_data_3 = all_data_3.transpose()
radius_3 = all_data_3[0]
mass_3 = all_data_3[1]


# No linear interpolation
# file_name = 'MvsR_old.dat'
# all_data = np.loadtxt(file_name,skiprows=2)
# all_data = all_data.transpose()
# radius = all_data[0]
# mass = all_data[1]

# The plot function can take two equally sized arrays,
# taking the first one as the x coordinates and the second one as the y coordinates
# plt.plot(radius, mass, '.', label = "Mass vs. Radius, Relativistic, No Interpolation")
plt.plot(radius_2, mass_2, '+', label = "Mass vs. Radius, Relativistic, Interpolation")
plt.plot(radius_3, mass_3, '+', label = "Mass vs. Radius, Relativistic, Long Phase change")

plt.xlabel('Stellar Radius $R$ (km)')
plt.ylabel('Stellar Mass $M/M_sol$')
# legend displays the labels given in each plot
plt.legend()
plt.show()

```

```python
plt.style.use('fivethirtyeight')

## Newtonian Calculations, M vs. R
# reading data stored in MvsR_Newtonian.dat

file_name = 'MvsR_Newtonian.dat'

# loadtxt reads all the data in a file and stores them in an array (assuming it's all floats)
# we use skiprows=1 to avoid the header
all_data = np.loadtxt(file_name,skiprows=1)

# We transpose the array in order to be able to extract each column individually
all_data = all_data.transpose()
radius = all_data[0]
mass = all_data[1]

# Each one of these is a numpy array that contains one of the columns in the .dat file.
# This will allow us to make operations on the extracted data
```

### Newtonian Calculations, $z$ vs. $M$

```python
plt.style.use('fivethirtyeight')

## Newtonian Calculations, z vs. M
# reading data stored in zvsM_Newtonian.dat

file_name = 'zvsM_Newtonian.dat'

all_data = np.loadtxt(file_name,skiprows=1)

all_data = all_data.transpose()
redshift = all_data[0]
mass = all_data[1]
```

```python
plt.plot(mass, redshift, label = "Gravitational Redshift vs. Mass, Newtonian")

plt.xlabel('Stellar Mass $M/M_sol$')
plt.ylabel('Gravitational Redshift $z$')
# legend displays the labels given in each plot
plt.legend()
plt.show()
```
