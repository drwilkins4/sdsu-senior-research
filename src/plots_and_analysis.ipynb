```python
# Importing mplot3d, numpy, and matplotlib
from mpl_toolkits import mplot3d
import numpy as np 
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')

## New model

file_name_1 = 'new_eq_of_state.dat'

# loadtxt reads all the data in a file and stores them in an array
# we use skiprows=1 to avoid the header
all_data_new = np.loadtxt(file_name_1,skiprows=2)

# We transpose the array in order to be able to extract each column individually
all_data_new = all_data_new.transpose()
energy_density_new = all_data_new[0]
pressure_new = all_data_new[1]

# Each one of these is a numpy array that contains one of the columns in the .dat file.


## Old model

file_name_2 = 'bag_model.dat'

all_data_old = np.loadtxt(file_name_2,skiprows=2)
all_data_old = all_data_old.transpose()
energy_density_old = all_data_old[0]
pressure_old = all_data_old[1]
```

```python
# taking the first one as the x coordinates and the second one as the y coordinates
plt.plot(energy_density_new, pressure_new, label = "Energy Density vs pressure, new Z bag model") # new
plt.plot(energy_density_old, pressure_old, label = "Energy density vs pressure, old bag model") # old

plt.xlabel('Energy Density $E_c$ (MeV/fm^3)')
plt.ylabel('Pressure $P_c$ (MeV/fm^3)')
# legend displays the labels given in each plot
plt.legend()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

file_1 = 'charge_neutrality.dat'

data_1 = np.loadtxt(file_1,skiprows=2)
data_1 = data_1.transpose()

iteration_1 = data_1[0]
charge_neutrality = data_1[1]


file_2 = 'chem_eq.dat'

data_2 = np.loadtxt(file_2,skiprows=2)
data_2 = data_2.transpose()

iteration_2 = data_2[0]
chemical_equilibrium_1 = data_2[1]  # mu_d - mu_u + mu_e = 0
chemical_equilibrium_2 = data_2[2]  # mu_d - mu_s = 0
```

```python
plt.plot(iteration_1, charge_neutrality) 

plt.xlabel('iteration')
plt.ylabel('charge')

plt.show()
```

```python
plt.plot(iteration_2, chemical_equilibrium_1, label='mu_d - mu_u + mu_e') 
plt.plot(iteration_2, chemical_equilibrium_2, label='mu_d - mu_s') 


plt.xlabel('iteration')
plt.ylabel('chemical equilibrium')

plt.legend()
plt.show()
```

```python
# Load files for chemical potential and number density plots


file_name_2 = 'chem_pot_vs_num_dens.dat'


all_data_2 = np.loadtxt(file_name_2,skiprows=2)


all_data_2 = all_data_2.transpose()


# Define data columns
mu_up = all_data_2[0]
mu_down = all_data_2[1]
mu_strange = all_data_2[2]
mu_electron = all_data_2[3]
number_density_up = all_data_2[4]
number_density_down = all_data_2[5]
number_density_strange = all_data_2[6]
number_density_electron = all_data_2[7]
total_number_density = all_data_2[8]
```

```python
## Chemical potentials vs individual number densities

plt.plot(total_number_density, mu_up, '+', label = "Up Quarks") 
plt.plot(total_number_density, mu_down, '.', label = "Down Quarks") 
plt.plot(total_number_density, mu_strange, label = "Strange Quarks")
plt.plot(total_number_density, mu_electron, label = "Electrons") 


plt.xlabel('Total Number Density (1/fm^3)')
plt.ylabel('Chemical Potential (MeV)')


plt.legend()
plt.show()
```

```python
## Individual number densities vs total number density

plt.plot(total_number_density, number_density_up/total_number_density, '+', label = "Up number density" )
plt.plot(total_number_density, number_density_down/total_number_density, '.', label = "Down number density" )
plt.plot(total_number_density, number_density_strange/total_number_density, label = "Strange number density" )
plt.plot(total_number_density, number_density_electron/total_number_density, label = "Electron number density" )

plt.xlabel('Total Number Density (1/fm^3)')
plt.ylabel('Individual Number Density / Total Number Density')


plt.legend()
plt.show()
```
